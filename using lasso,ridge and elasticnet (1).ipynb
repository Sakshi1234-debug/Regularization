```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression, Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#load the diabetes dataset
d=load_diabetes()
data=pd.DataFrame(data=d.data,columns=d.feature_names)
data.head()
```

```python
data['Target']=d.target
data.shape
```

```python
data.head()
```

```python
#select features and target
x=data.iloc[:, :-1]
y=data['Target']
#split the data into training and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
```

```python
l=LinearRegression()
#fit the model on training data set
l.fit(x_train,y_train)
#predict using linear regression
y_linear_pred_train=l.predict(x_train)
y_linear_pred=l.predict(x_test)
```

```python
#create ridge model
r=Ridge(alpha=1.0)
#fit the model
r.fit(x_train,y_train)
#predict using ridge regression
y_ridge_pred_train=r.predict(x_train)
y_ridge_pred=r.predict(x_test)
```

```python
#create lasso model
la=Lasso(alpha=1.0)
#fit the model
la.fit(x_train,y_train)
#predict using lasso  regression
y_lasso_pred_train=la.predict(x_train)
y_lasso_pred=la.predict(x_test)
```

```python
#create ElasticNet model
e=ElasticNet()
#fit  the model
e.fit(x_train,y_train)
#predict using elastic net
y_elasticnet_pred_train=e.predict(x_train)
y_elasticnet_pred=e.predict(x_test)
```

```python
#create mean squared errors for all  three models
#linear
mlt=mean_squared_error(y_train,y_linear_pred_train)
ml=mean_squared_error(y_test,y_linear_pred)
print('Linear Regression Results','\n')
print('Training error : ',mlt)
print('Testing error : ',ml)
print('Difference :',ml-mlt)
print('---------------------------------------------------')
#Ridge
mrt=mean_squared_error(y_train,y_ridge_pred_train)
mr=mean_squared_error(y_test,y_ridge_pred)
print('Ridge Regression Results','\n')
print('Training error : ',mrt)
print('Testing error : ',mr)
print('Difference :',mr-mrt)
print('---------------------------------------------------')
#lasso
mlat=mean_squared_error(y_train,y_lasso_pred_train)
mla=mean_squared_error(y_test,y_lasso_pred)
print('Lasso Regression Results','\n')
print('Training error : ',mlat)
print('Testing error : ',mla)
print('Difference :',mla-mlat)
print('---------------------------------------------------')
#ElasticNet
met=mean_squared_error(y_train,y_elasticnet_pred_train)
me=mean_squared_error(y_test,y_elasticnet_pred)
print('ElasticNet Regression Results','\n')
print('Training error : ',met)
print('Testing error : ',me)
print('Difference :',me-met)
print('---------------------------------------------------')
```
